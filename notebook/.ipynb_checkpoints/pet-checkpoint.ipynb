# View PET image

```python
import ants
import matplotlib 
import subprocess
import numpy as np
```

#### PET images are in DICOM (.dcm) format. There is a tool (dcm2niix) which will convert to NIfTY format (nii)

```python
if subprocess.call(['dcm2niix', 'dicom_pet']): # We convert or print error message
    print('Error converting file from DICOM to NIfTY')
```

Load image

```python
img = ants.image_read('dicom_pet/dicom_pet_Brain-FDG_20070801103000_0.nii')
```

Plot all axis

```python
getting_wild = matplotlib.cm.get_cmap('magma')
for ax in [0, 1, 2]:
    img.plot(axis=ax, ncol=6, cmap=getting_wild)
```
