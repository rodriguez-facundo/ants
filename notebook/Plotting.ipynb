# Exploring plot capabilities

```python
import ants
import numpy as np
import matplotlib
```

Import mni atlas

```python
mni = ants.get_data('mni')
```

Create a ant image from the data

```python
img = ants.image_read(mni)
```

Plot

```python
img.plot()
```

This is 3D data, so lets plot another axis

```python
img.plot(axis=2)
```

Change grid

```python
img.plot(axis=1, ncol=6)
```

Lets put some color

```python
getting_wild = matplotlib.cm.get_cmap('plasma')
```

```python
img.plot(axis=1, ncol=6, cmap=getting_wild)
```

Lets compare 2 volumetric images (pretending we already segmented one volume)

```python
oasis_mni = ants.get_data('oasis_mni') # This is a segmented atlas
oasis_img = ants.image_read(oasis_mni)
```

```python
ants.plot(img, oasis_img)
```

```python
ants.plot(img, oasis_img, overlay_alpha=0.2)
```
